<h2 style="color:#621940">SURVEY</h2>

Unsupervised machine learning is a machine learning algorithm that is used to draw inferences from data without supervision i.e. without labeled responses. In this section, research that uses unsupervised learning for anomaly detection in epileptic EEG signals is briefly highlighted.



<h3 style="color:#843b62;">Unsupervised automatic seizure detection for focal-onset seizures recorded with behind-the-ear EEG using an anomaly-detecting generative adversarial network</h3>

<h3>Introduction</h3>

<b>Epilepsy</b> is a neurological disorder that disrupts brain activity and is characterized by seizures involving momentary loss of consciousness or uncontrolled jerking movements. The consequences of seizures stemming from epilepsy lower the quality of life for patients as well as caregivers. Accurate detection and monitoring of seizures is therefore necessary for patient diagnosis and patient management.

Some patients are unable to recognize the occurrence of seizures without external monitoring. According to Hoppe et al., 55% of reported seizures are unrecognized by patients thus long-term monitoring in daily life is imperative for seizure detection. The conventional method adopted for the diagnosis of epilepsy is <span style="color:#f67e7d; font-weight:bold;">video-electroencephalogram (EEG)</span> monitoring. This is done using scalp EEG or intracranial-EEG and both methods are impractical for long-term monitoring in daily life. For this reason, the use of <b>behind-the-ear EEG</b> which involves the use of ear electrodes for recording is encouraged.

Studies have shown that EEG recorded behind the ear is comparable to scalp EEG. This is corroborated using many EEG-related tasks such alpha attenuation, the auditory steady-state response and steady-state visually-evoked potentials.

Long-term monitoring of patients generates a lot of data and this data can be used to automatically detect seizures with the help of machine learning algorithms. In this study, an <span style="color:#f67e7d; font-weight:bold">anomaly generative adversarial network(AnoGAN)</span> is trained to the learn the <i><b>time-frequency variance</b></i> in the spectral images of behind-the-ear EEG records. The model is subsequently used to detect ictal events in epilepsy patients.

<br/>

<h3>Methodology</h3>
<h4>Subjects</h4>

The study involved patients who had undergone epilepsy surgery following intracranial EEG recording at the Samsung Medical Center, Seoul, Korea from 2016 to 2019. It included 5 males and 7 females whose mean age at seizure onset was 8.6 + 6.7 years (range: 1-27 years). At the time of intracranial EEG examination, the mean age of patients was 23.8 + 5.34 years(range: 13-32 years). All the patients had undergone noninvasive evaluations which included 3 Tesla brain MRIs(Magnetic Resonance Image) and neuropsychological examination.

<img src="Subjects.png" alt="Image of patients" />

<br/>

<h4>Recording EEG</h4>

Video-EEG monitoring in the epilepsy monitoring unit began following the implantation of intracranial electrodes. EEGs were recorded by the intracranial electrodes at the presumed seizure onset zone. Behind-the-ear electrodes were also placed to simultaneously record data. EEGs were recorded using NicoletOne EEG system with a sampling frequency of 512Hz. The recorded data were subsequently reviewed by two epileptologists, who determined and marked the onsets and ends of seizures based on intracranial EEG. Simultaneously recorded EEG data using behind-the-ear electrodes were then analyzed using a GAN.

The patients were categorized under training and test groups and both groups had 6 patients. The training set was created by extracting the non-ictal state recording of the training group for at least 60 minutes before the first ictal event. The training set was then used to train a GAN to learn the representation of normal states with unsupervised learning, while the normal and ictal recordings from the test group was used to validate the performance of the seizure detection.

<br/>

<h4>Data preprocessing</h4>

Behind-the-ear signals were filtered using a 16th order <i>Butterworth bandpass filter</i> with a high-pass cutoff frequency of 0.5Hz and a low-pass cutoff frequency of 50Hz. The behind-the-ear signals were transformed to images by firstly calculating the power spectral density (PSD) with the short-time Fourier transform. A PSD image was constructed for each behind-the-ear channel by applying short-time Fourier transform for every 32-s window with 50% overlap.

Before feeding the images into the GAN for training, EEG spectrograms were normalized using z-score normalization and scaled within a range of (-1,1) for each channel.Two spectrogram images were formed for each behind-the-ear EEG channel. Finally, the PSD image and the two spectrograms for each channel were stacked to form a data structure of $N \times 64 \times 64 \times 3$. 

<br/>

<h4>Training the Anomaly generative adversarial network.</h4>

The GAN was finally constructed from a <b>deep convolutional generative adversarial network (DCGAN)</b>. A DCGAN is a GAN with convolutional layers, which stabilize the adversarial learning process and improve the capacity of learning the visual information. A DCGAN has two adversarial structures, a generator G, and a discriminator, D. G creates the data from a random vector in latent space z, while D compares the generated data with a real-world samples to assess its genuity. During the training, G learns a mapping function $G(z)$ from the latent space $z$ to the image $x$. In this case, $z$ is a random 1-dimensional vector from the latent space $z$ which contains condensed information of the spectrogram images.

The GAN was first trained to learn the normal time-frequency variance in the spectrogram images. The input of the GAN was spectrogram images of behind-the-ear EEG. Both the generator and discriminator used 4 transposed convolutional layers with a
$5 \times 5$ kernel as shown in the figure below. Batch normalization was applied to each convolutional layer to stabilize and accelerate training.

<img src="GAN.png" alt="image of GAN" />

The GAN was trained for 100 epochs with a learning rate of 0.0002. A stochastic gradient descent optimizer for the discriminator whereas an Adam optimizer was used for the generator. The loss of the discriminator was a sigmoid cross-entropy for distinguishing the generated samples from the actual samples while the loss of the generator was a sigmoid-entropy for deceiving the discriminator with generated samples from a latent vector of $z$. 

The GAN is a generative and unsupervised model which makes performance measurement a challenging task. In general GANs are usually evaluated with the visual observation of generated images in the computer vision domain. This GAN is however trained to learn the variability of the PSD from the behind-the-ear EEG so a visual observation of the generated image is not appropriate. Thus, stabilizing the loss of network i.e. discriminative loss and generative loss during training was very imperative.

<br/>

<h4>Anomaly Detection</h4>

Anomalies were identified by using the GAN that was trained on the normal dataset as a mapping function from the latent space. As aforementioned, the generator learned a mapping function $G(z) = Z \Longrightarrow X$ from the latent space representation $z$, to real-domain images $X$ via adversarial training with the normal dataset. With the trained generator network, the latent vector that produced an image closest to the test image was found. The anomaly loss measure was used to find the similarity between the generated image $G(z)$, and the test image $X$. The latent vector that produced the lowest anomaly loss for each image generated was also found.

<b>Anomaly loss equation</b>

<p style="font-size: 2rem; text-align: center;">$L_{anomaly} = (1 - \lambda) * L_{residual}(z) + \lambda*L_{discrimination}(z)$</p>

<p style="font-size: 2rem; text-align: center;">$L_{residual} = \sum|x - G(z)|$</p>

<p style="font-size: 2rem; text-align: center;">$L_{discrimination} = \sum|D(x) - D(G(z))|$</p>

In this study, the lambda parameter was empiracally selected to be $0.5$.

Using this anomaly loss, when the test input image was from the normal state, the generated image $G(z)$ and input image $X$ were similar however for the abnormal state(i.e. ictal event), the generated image $G(z)$ and input image were dissimilar thus the network was able to detect ictal events as an abnormality.

The gram matrix, also known as the style loss was later combined with the anomaly loss to improve the performance of anomaly detection.


<br/>

<h3>Result</h3>

<h4>Detection performance</h4>

The algorithm was evaluated with test data from 6 patients as previously mentioned. The performance of the algorithm is described in the table below. The algorithm shows $0.9372$ as the <i>mean area under the receiver operating characteristic <b>(AUROC)</b></i> without the gram matrix and attained an AUROC of $0.9393$ by combining the anomaly score with the gram matrix score.

<h2 style="color: #843b62">The K-Means Clustering algorithm</h2>

The K-Means algorithm is one of the unsupervised learning methods that is used to group objects by <b>similarity</b>. The K-means algorithm belongs to the category of prototypical clustering which means that each cluster is represented by a prototype such a centroid(mean) for continuous attributes or medoid(mode) in the case categorical features.

<b>What exactly is the K-means algorithm?</b>

In this section, the algorithm is summarized in 4 steps:

1. Randomly pick k centroids from the sample points as initial cluster centers.
2. Assign each sample point to the nearest centroid $\mu^{j}, j \in {1, ..., k}$
3. Move to centroid to the center of the samples that were assigned to it.
4. Repeat steps 2 and 3 until the cluster assignments do not change or a user-defined tolerance or maximum number of iterations is reached.

<b>Why use the K-Means clustering algorithm?</b>

- It is very computationally efficient.
- It is easy to implement.

<br/>

<h3 style="color:#621940">Similarity between sample points</h3>

The effectiveness of a clustering method depends on the similarity metric used and this is strongly problem-dependent. In this application, the traditional Euclidean distance will be used as the similarity metric because the data will be standardized before training.

<p style="font-size: 2rem; text-align: center;">$\delta(x, y) = \sum^{m}_{j=1}(x_{j} - y_{j})^{2} = \parallel x - y \parallel ^{2}_{2}$</p>

<b>Where:</b>

<i><strong>j</strong> = jth dimension (feature column) of the sample points</i> ($x$ and $y$)

<br/>

Using the Euclidean distance as a similarity metric, the K-means algorithm can be viewed as a simple optimization problem where an iterative approach is used to minimize the within-cluster distance also known as the Sum of Squared Errors (SSE).

<p style="font-size: 2rem; text-align: center;">$SSE = \sum^{n}_{i=1}\sum^{k}_{j=1}w^{i, j}\parallel x^{(i)} - \mu^{(j)}\parallel^{2}_{2}$</p>

<b>Where: </b>

$\mu^{(j)}$ = centroid for cluster j

$w^{i, j}$ = 1 if sample $x^{i}$ is in cluster j; 0 otherwise

$i$ = sample index

<br/>

<h3 style="color: #843b62">K-Means Model evaluation with Silhouette plots</h3>

Quantifying the quality of clustering using Silhouette Plots

Silhouette Analysis can be used as a graphical tool to plot and measure how tightly coupled grouped samples are.

The following steps are used to calculate the silhouette coefficient of each sample

- Calculate cluster cohesion $a^{(i)}$ as the average distance between a sample $x^{(i)}$ and all other points in the same cluster.

- Calculate the cluster separation $b^{(i)}$ from the next closest cluster as the average distance between the sample 
$x^{(i)}$ and all samples in the nearest cluster.

- Calculate the silhouette $s^{(i)}$ as the difference between the cluster cohesion and cluster separation divided by the greater of the two.


<p style="font-size: 3rem; text-align: center;">$s^{(i)} = \frac {b^{(i)} - a^{(i)}} {max(b^{(i)}, a^{(i)})}$</p>

For an ideal model, the silhouette ratio for each point should be 1.

```python
from matplotlib import cm
from sklearn.metrics import silhouette_samples

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(X_train_std, y_km, metric="euclidean")

```

```python
cluster_labels + 1
```

```python
y_ax_lower, y_ax_upper = 0, 0
yticks = []

fig, ax = plt.subplots()

for idx, cluster in enumerate(cluster_labels):
    cluster_silhouette_vals = silhouette_vals[y_km == cluster]
    cluster_silhouette_vals.sort()
    y_ax_upper += len(cluster_silhouette_vals)
    color = cm.jet(float(idx) / n_clusters)
    ax.barh(range(y_ax_lower, y_ax_upper), cluster_silhouette_vals, height=1.0, edgecolor='none', color=color)
    yticks.append((y_ax_lower + y_ax_upper) / 2)
    y_ax_lower += len(cluster_silhouette_vals)
    
silhouette_avg = np.mean(silhouette_vals)
ax.axvline(silhouette_avg, color="red", linestyle="--")
plt.yticks(yticks, ["B", "A"])
ax.set_ylabel("Cluster")
ax.set_xlabel("Silhouette coefficient")
plt.show()
```

```python
print("Silhouette avg: {}".format(silhouette_avg))
```

<h2 style="color: #621940">PCA (Principal Component Analysis)</h2>

PCA is an unsupervised dimensionality reduction method that is used to extract features that maximize variance within a dataset. It is a technique that compresses data while maintaining most relevant information.

<br/>

<h3 style="color:#843b62">How does PCA compress data?</h3>

PCA aims to find the directions of maximum variance in high-dimensional data and projects it onto a new subspace with equal or fewer features than the original. The orthogonal axes (principal components) of the new subspace can be interpreted as the directions of maximum variance with the constraint that the axes are orthogonal to each other as shown in the figure below.

<img alt="PCA" src="pca.png" />

Using PCA for dimensionality reduction involves constructing a $d x k$ - dimensional matrix $W$ which maps a sample vector $x$ onto a new k-dimensional feature space with fewer dimensions the original d-dimensional feature space.

<p style="text-align: center; font-size: 2rem">$x = [x_{1}, x_{2}, ...., x_{d}], x \in \mathbf{R}^{d}$</p>

<p style="text-align: center; font-size: 2rem">$\downarrow xW, W \in  \mathbf{R} ^{d x k}$</p>

<p style="text-align: center; font-size: 2rem">$z = [z_{1}, z_{2}, ....., z_{k}], z \in \mathbf{R}^{k}$</p>

It is worthy to note that PCA directions are <strong>highly sensitive to data scaling</strong> so there's a need to scale the data in order to assign equal importance to all features.

<br/>

<h3 style="color: #843b62">The Algorithm</h3>

1. Standardize the d-dimensional dataset.
2. Construct the covariance matrix.
3. Decompose the covariance matrix into its eigenvectors and eigenvalues.
4. Sort the eigenvalues in decreasing order to rank its eigenvectors.
5. Select k eigenvectors which correspond to the k largest eigenvalues, where k is dimensionality of the new feature subspace $(k\leq d)$
6. Construct a projection matrix $W$ from the top k eigenvectors.
7. Transform the d-dimensional input dataset $X$ using the project matrix $W$ to obtain the new $k$-dimensional feature subspace.

<br/>

<h3 style="color:#843b62">Extracting the Principal components using Scikit-learn</h3>

Unlike K-Means clustering, PCA does not predict the membership a sample point to a cluster. A supervised learning model can be trained using the new principal components. 

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

columns = [column for column in list(ep_df.columns) if column != "Case ID" if column != "Label"]

ep_df2 = ep_df[columns] # X
y = ep_df["Label"].map({"N" : 0, "S": 1})

# splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(ep_df2, y, test_size=0.3, random_state=0)

# standardization (Z-score normalization)
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)


pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)

```

<h3 style="color:#843b62">Plotting decision regions</h3>

```python
from matplotlib.colors import ListedColormap
def plot_decision_regions(X, y, classifier, resolution=0.02):
    
    # marker generator and color map
    colors = ('red','lightgreen')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
    
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[ y == cl, 1], alpha=0.6, \
                    c=cmap(idx), edgecolor='black', marker="o", label=cl)
```

```python
plot_decision_regions(X_train_pca, y_train, classifier=svm)
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.legend(loc="lower left")
plt.title("Training data")
plt.show()
```

```python
plot_decision_regions(X_test_pca, y_test, classifier=svm)
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.legend(loc="lower left")
plt.title("Test data")
plt.show()
```

```python
from sklearn.metrics import accuracy_score
y_pred = svm.predict(X_test_pca)
print("Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred) * 100))
```
